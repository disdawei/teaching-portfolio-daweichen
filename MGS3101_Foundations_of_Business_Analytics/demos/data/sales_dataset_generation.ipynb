```python
import numpy as np
import pandas as pd
np.random.seed(42)
```

```python
# Create a monthly date index
dates = pd.date_range(start="2024-01-01", periods=12, freq="MS")

# Simulate monthly sales (base + random noise)
base = 200
trend = np.linspace(0, 60, 12)  # simple upward trend
noise = np.random.normal(0, 10, 12)
sales = base + trend + noise

# Wrap into a pandas Series for convenience
sales_series = pd.DataFrame({'Date': dates, 'Sales A': sales})
sales_series
```

```python
noise_b = np.random.normal(0, 12, 12)
sales_b = base + trend * 0.6 + 25 + noise_b
sales_series['Sales B'] = sales_b
sales_series
```

```python
sales_series.to_csv('sales.csv', index=False)
```
