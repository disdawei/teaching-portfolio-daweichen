## 3. Two Ways to Create Plots in Matplotlib

Matplotlib provides two main interfaces for plotting:

1. **[Pyplot (state-based) interface](https://matplotlib.org/stable/api/pyplot_summary.html)** → using `plt.plot()`

2. **[Object-oriented (OO) interface](https://matplotlib.org/stable/api/axes_api.html)** → using `fig` and `axes` objects

They often produce the same visual result, but differ in **control**, **scalability**, and **recommended use**.

### 3.1 The Pyplot Interface `plt.plot()`

**Pros:**
- Quick and simple — good for small, one-off plots.

- matplotlib automatically: (1) Creates a Figure (if none exists). (2) Adds Axes (a plotting area). (3) Draws your plot on the current Axes.

- You don’t manually handle figure or axes objects.

**Cons**

- Harder to manage when you have multiple subplots or need fine control.

- The “current figure/axes” concept can get confusing in larger scripts.

### 3.2 The Object-Oriented Interface `ax.plot()`

1. Creat a **Figure** object using `fig = plt.figure()` — think of it as the **drawing board** where all visualizations will be placed.

2. Add **Axes** using `ax = fig.add_subplot(111)` — create the specific region — or “**drawing paper**” — within the **Figure** where our plots will appear.

- A single **Figure** can contain multiple **Axes (or subplots)**, each functioning as an independent plotting area.

## 8. Common Pitfalls to Avoid

- Truncated or inconsistent axes that exaggerate differences  
- Over-cluttered visuals (too many lines, labels, or effects)  
- Missing context: unlabeled axes, no units, no legend  
- Hard-to-read date ticks (overlapping labels)  
- Not disclosing smoothing or data cleaning steps
