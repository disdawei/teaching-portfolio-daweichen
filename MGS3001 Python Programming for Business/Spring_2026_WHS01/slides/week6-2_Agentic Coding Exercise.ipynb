## 1. Recap: The 4D AI Fluency Framework

### 1-1. Quick Recap

<img src="../demo/4D.png" alt="4D AI Fluency Framework" width="500">

- **Delegation** — What should AI do, and what should humans do?
  - **Problem awareness:** define the goal clearly before asking AI to help
  - **Platform awareness:** know which AI tools are available and suitable
  - **Task delegation:** divide work intentionally between humans and AI

- **Description** — How clearly are the task, context, and expected output defined and communicated with AI?
  - **Product:** define the desired output
  - **Process:** guide AI on the approach when needed
  - **Performance:** specify expectations for style, clarity, or difficulty level

- **Discernment** — Is the output useful, correct, and appropriate?
  - **Product:** evaluate output quality (what aspects?)
  - **Process:** evaluate the problem-solving approach
  - **Performance:** evaluate AI behavior and limitations

- **Diligence** — Is AI used responsibly, with ownership of the results?
  - **Creation:** be intentional about which AI systems to use and how
  - **Transparency:** be open about AI involvement when relevant
  - **Deployment:** take responsibility for outputs you use or share

> **Reference:** [AI Fluency: Framework & Foundations](https://anthropic.skilljar.com/ai-fluency-framework-foundations) by Anthropic

## 3. GitHub and Project Workflow

So far, we have focused on using **VS Code**, **Jupyter Notebook**, and **AI** to complete a task.  
The next step is to place that work into a more complete project workflow by introducing **GitHub**.

A simple way to think about GitHub is this:

**GitHub is like Google Drive for code — but it also keeps the history of every change you make.**

In this course, GitHub will be useful in two main ways:

- to save, organize, track, and share your own Python project files
- to find useful examples or starter projects shared by others

The goal today is **not** to master GitHub.  
The goal is to understand why it matters and how it connects with your future Python project workflow.

### 3-1. What Is GitHub, and Why Should You Care?

GitHub is a platform for storing, organizing, and sharing coding projects.

For beginners, its value is not only that it stores files online, but that it helps you keep your work more organized over time.

For example, instead of managing many files such as:

- `project.ipynb`
- `project_final.ipynb`
- `project_final_v2.ipynb`
- `project_final_real_final.ipynb`

GitHub helps you keep a clearer and more structured history of your project.

You can think of this as somewhat similar to **Track Changes in Word** or saving named versions of a report, except GitHub is designed for code and project files.

This matters because real project work often involves:

- revising analysis
- updating files over time
- returning to earlier versions
- building on previous work
- keeping a record of how a project developed

So even if you are new to coding, GitHub is useful because it helps you manage project work more professionally.

**Why upload it?**

- so you can save your work in a more organized way
- so you can revisit or improve it later
- so you can start building a project history
- so your coding work is not stored only on one computer

A useful question to ask is:

**If I return to this project two weeks later, will I still know which file is the latest and what I changed?**

GitHub helps solve that problem.

### 3-5. How to Read and Understand a Repository

When you open a repository, do not think of it as something to copy immediately.

First, try to understand what it is.

A good place to start is:

- the **README**
- the main files and folders
- any notebook or main script
- the instructions for running the project

Ask simple questions such as:

- What is this project trying to do?
- What file should I open first?
- Does this look manageable for me?
- What parts do I already understand?
- What parts would I need help with?

This is also a good moment to use AI well:

- ask AI to explain the repository structure
- ask AI to summarize what a script does
- ask AI which file seems most important
- ask AI to explain unfamiliar code in beginner-friendly language

### 3-6. Fork, Clone, and Working Locally

Later, when you want to actually use a repository, you will usually need your own version of it.

There are two common ideas:

- **Fork**: create your own copy on GitHub
- **Clone**: download a copy to your computer

You do not need to master these terms today.

For now, the main idea is simple:

**GitHub stores projects online, and VS Code is where you work on them locally.**

That connection will become more important in later sessions.

### 3-7. Open and Work on the Project in VS Code

Once a project is on your computer, you can open it in **VS Code**.

This is where the workflow comes together:

- **GitHub** stores the project
- **VS Code** is where you read and edit files
- **Jupyter Notebook** helps you test code step by step
- **AI tools** help you explain, revise, and debug code

At this stage, your job is not to rewrite everything from scratch.

Instead, focus on:

- understanding the project structure
- identifying the main files
- making small changes
- testing those changes
- using AI to support your understanding

### 3-8. Version Control in Plain Language

One of the most important ideas behind GitHub is **version control**.

That phrase may sound technical, but the idea is simple:

**Version control means keeping an organized history of your project so you can see what changed, when it changed, and return to earlier versions if needed.**

For business students, you can think of it like this:

Instead of managing files such as:
- `draft_v1`
- `draft_v2`
- `draft_final`
- `draft_final_revised`
- `draft_final_revised_REAL`

GitHub keeps the history in a more structured way.

This becomes useful when you:

- improve a notebook over time
- try a change and want to undo it
- compare an earlier and later version
- keep your project work more organized

You do not need to learn every Git command today.  
The important point is to understand why project history matters.

### 3-9. Key Takeaway

GitHub is not just a place to download code. It is a practical tool for saving, organizing, revising, and building on project work more professionally.
