## 3. Opening Files

### 3-1. `open()` and File Modes

**General syntax:**

```python
open(file_path, mode)
```

**Common modes:**

- `"r"` → Read (default): opens a file for reading, raises an error if the file does not exist
- `"w"` → Write: opens a file for writing, creates a new file if it does not exist or overwrites an existing one
- `"a"` → Append: opens a file to add content at the end, creates the file if it does not exist
- `"x"` → Create: creates a new file, raises an error if the file already exists
- `"t"` → Text model (default value)
- `"b"` → Binary mode (e.g. images, non-text files)

***Note:** Modes can be combined, e.g., `"rb"` (read binary), `"wt"` (write text).*

For more details, refer to [the official Python documentation](https://docs.python.org/3/library/functions.html#open) for the built-in `open()` funciton. Focus on **parameters** and **return values**.


### 3-2. Basic Example: Open a File in Read Mode

This is the traditional style:

```python
f = open("data.txt", "r")
print(f.read())
f.close()
```


````python
# Basic example: open a file in read mode
f = open("data.txt", "r", encoding="utf-8")
print(f.read())
f.close()
````

**Think:** What could happen if we forget to close the file?

It has a weakness:

- you must remember to call `close()`
- forgetting to close files can waste resources and make code harder to manage

That is why Python programmers usually prefer `with open(...) as ...`.

### 3-3. Best Practice: `with open(...) as f`

A common and safer way to work with files is:

```python
with open("data.txt", "r") as f:
    content = f.read()
```

What this means:

- `open("data.txt", "r")` opens the file in read mode
- `with ... as ...` creates a temporary context for working with the file
- `f` is a variable that refers to the opened file object

Why use `with`?

- the file is closed automatically after the block ends
- the code is cleaner
- it is safer in larger programs

`f` has no special meaning by itself. It is simply a short variable name. You could also write `as file` instead.


````python

# Recommended style: with open(...)
with open("data.txt", "r", encoding="utf-8") as f:
    content = f.read()

print(content)
````

## 4. File Paths

### 4-1. Why File Paths Matter

When we write:

```python
open("data.txt", "r")
```

How does Python know where `data.txt` is?

To answer that, we need to understand **file paths**.


### 4-2. Current Working Directory (CWD)

The current working directory is the **folder** where Python is currently running.

When you use a **relative path**, Python *looks for the file starting from this folder*.


````python

import os

print("Current working directory:")
print(os.getcwd())
````

### 4-3. Absolute vs Relative Paths

**Absolute path:** The *full location* of a file on your computer.

Example:

```python
"C:/Users/YourName/project/data/data.txt"
```

**Relative path:** A path written *relative to the current working directory*.

Examples:

```python
"data.txt"
"data/sales.txt"
```

### 4-4. Parent and Child Directories

Relative paths allow you to navigate around the current working directory:

- Current directory: `data.txt` or `./data.txt`

- Child directory (subfolder): `data/sales.txt` or `./data/sales.txt`

- Parent directory:  `../notes.txt`

- Another folder in the parent directory: `../data/sales.txt`

Common symbols:

- `.` → current directory  
- `..` → parent directory  

````python

# Relative path example: file in a child folder
with open("data/sales.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

### 4-5. Why the Current Working Directory Matters

A very common error is:

```python
FileNotFoundError
```

This often happens because:

- the file is not in the current working directory
- the filename is wrong
- the path is wrong

➡️ Many file errors are really **path errors**.

Why separate folders?

## 5. Reading Text Files

### 5-1. Common Ways to Read a File

Common methods:

- `read()` → read the whole file as one string
- `readline()` → read one line at a time
- `readlines()` → read all lines into a list

You can also loop over the file directly.


````python

# read()
with open("data.txt", "r", encoding="utf-8") as f:
    print("Using read():")
    print(f.read())
````

````python

# readline()
with open("data.txt", "r", encoding="utf-8") as f:
    print("Using readline():")
    print(f.readline().strip())
    print(f.readline().strip())
````

````python

# readlines()
with open("data.txt", "r", encoding="utf-8") as f:
    print("Using readlines():")
    print(f.readlines())
````

````python

# loop over lines
with open("data.txt", "r", encoding="utf-8") as f:
    print("Looping over the file:")
    for line in f:
        print(line.strip())
````

### 5-2. Processing Structured Text Data

Even though `.txt` files are plain text, they can still contain structured information.

For example, this file:

```text
Alice,100
Bob,200
Alice,150
Cindy,250
```

looks similar to a simple CSV-style dataset.

➡️ Files store raw text, but Python can turn that text into structured data.


````python

sales_totals = {}

with open("data/sales.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, amount = line.strip().split(",")
        amount = int(amount)
        sales_totals[name] = sales_totals.get(name, 0) + amount

print(sales_totals)
````

## 6. Writing and Appending Files

Two very important modes are:

- `"w"` → write from the beginning; old content is overwritten
- `"a"` → append to the end; old content stays

This difference is very important in practice.


````python

# 'w' overwrites the file
with open("results/example_write.txt", "w", encoding="utf-8") as f:
    f.write("First version\n")

with open("results/example_write.txt", "w", encoding="utf-8") as f:
    f.write("Second version\n")

with open("results/example_write.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

````python

# 'a' appends to the file
with open("results/example_append.txt", "w", encoding="utf-8") as f:
    f.write("Line 1\n")

with open("results/example_append.txt", "a", encoding="utf-8") as f:
    f.write("Line 2\n")
    f.write("Line 3\n")

with open("results/example_append.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

## 7. Handling Errors

### 7-1. File Work Can Fail

File operations are uncertain. Files may not exist, paths may be wrong, and data may not be clean.

So we often use `try` and `except`.


````python

try:
    with open("data/missing_file.txt", "r", encoding="utf-8") as f:
        print(f.read())
except FileNotFoundError:
    print("The file was not found.")
````

### 7-2. Handling Messy Data

Real-world data is often imperfect.

For example, `data/messy_sales.txt` contains some bad lines:

```text
Alice,100
Bob,200
Alice-150
Cindy,250
David,not_available
```

A robust program should skip invalid lines instead of crashing.


````python

totals = {}

with open("data/messy_sales.txt", "r", encoding="utf-8") as f:
    for line in f:
        try:
            name, amount = line.strip().split(",")
            amount = int(amount)
            totals[name] = totals.get(name, 0) + amount
        except ValueError:
            print("Skipped invalid line:", line.strip())

print("Clean totals:", totals)
````

## 8. Using Modules

### 8-1. What Is a Module?

A **module** is a single Python file that contains reusable code (functions, variables, etc.) that you can import into your program.

Modules help us:
- avoid rewriting common functions  
- organize code into smaller components  
- work more efficiently  

Related Concepts

- *Module* → one Python file  
  Example: `math`, `os`, `json`

- *Package* → a collection of related modules  
  Example: `pandas` (contains many modules)

- *Library* → a broader term for a set of packages/modules used together  
  Example: the Python standard library

Examples

- `math` → mathematical functions  
- `os` → interacting with the operating system  
- `json` → working with JSON data  
- `pandas` → data analysis (a package with multiple modules)


### 8-2. Different Ways to Import Modules

Three common ways:

1. *Import the whole module*
```python
import math
math.sqrt(16)
```

2. *Import a specific function*
```python
from math import sqrt
sqrt(16)
```

3. *Import with an alias*
```python
import pandas as pd
```

Different styles affect readability and convenience.


````python

import math

print(math.sqrt(16))
print(math.ceil(3.2))
print(math.floor(3.8))
````
