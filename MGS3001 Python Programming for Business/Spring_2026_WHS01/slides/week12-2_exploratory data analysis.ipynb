## What Is EDA?

**Exploratory Data Analysis** is the systematic examination of a dataset to understand its **main characteristics** before applying formal statistical methods.

The core idea: <span style="color:red">**let the data speak before imposing assumptions.**</span>

### Three Questions EDA Answers
*`Note: The GitHub repository dataset will be used for illustration purposes.`*

| Question | What You Do | Example |
|---|---|---|
| **What does my data look like?** | Summary statistics, distributions | Mean stars = 61,537 with heavy right skew |
| **Are there patterns or relationships?** | Correlations, group comparisons | Repos by organizations have more stars than individual repos |
| **Are there problems to address?** | Outliers, skewness, missing patterns | One repo has 685K stars — outlier or genuine variation? |

### Skewness

Many real-world count variables are **right-skewed**: most values are small, but a few are very large.

- Skewness ≈ 0: roughly symmetric
- Skewness > 1: right-skewed (long tail on the right)
- Skewness < -1: left-skewed (long tail on the left)

<img src="skewed_distributions.svg" width="80%">

```python
# Check skewness for numeric variables
skew_df = df[num_vars].skew().round(2).to_frame("skewness")
skew_df["interpretation"] = skew_df["skewness"].apply(
    lambda x: "Roughly symmetric" if abs(x) < 1
    else ("Right-skewed" if x > 0 else "Left-skewed")
)
skew_df
```

### Log Transformation for Skewed Variables

When a variable is heavily right-skewed, **a log transformation can make the distribution more symmetric**. This is important because:
1. Many statistical methods assume roughly normal distributions
2. Log-transformed coefficients have a useful interpretation: a 1-unit change in the predictor is associated with a **percentage** change in the outcome

We use `np.log1p()` (= log(x + 1)) instead of `np.log()` to handle zeros safely.

```python
# Before vs. After log transformation — stargazers_count
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Before
axes[0].hist(df["stargazers_count"].dropna(), bins=20, edgecolor="white", color="#3B82F6")
axes[0].set_title("stargazers_count (raw)", fontweight="bold")
axes[0].set_xlabel("Stars")
axes[0].set_ylabel("Frequency")
axes[0].annotate(f"Skewness: {df['stargazers_count'].skew():.2f}",
                 xy=(0.65, 0.85), xycoords="axes fraction", fontsize=11,
                 bbox=dict(facecolor="lightyellow", edgecolor="gray", boxstyle="round"))

# After
log_stars = np.log1p(df["stargazers_count"])
axes[1].hist(log_stars.dropna(), bins=20, edgecolor="white", color="#16A34A")
axes[1].set_title("log(stargazers_count + 1)", fontweight="bold")
axes[1].set_xlabel("log(Stars + 1)")
axes[1].set_ylabel("Frequency")
axes[1].annotate(f"Skewness: {log_stars.skew():.2f}",
                 xy=(0.65, 0.85), xycoords="axes fraction", fontsize=11,
                 bbox=dict(facecolor="lightyellow", edgecolor="gray", boxstyle="round"))

plt.suptitle("Effect of Log Transformation", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()
```

```python
# Create log-transformed variables for skewed counts
df["log_stars"] = np.log1p(df["stargazers_count"])
df["log_forks"] = np.log1p(df["forks_count"])
df["log_issues"] = np.log1p(df["open_issues_count"])

print("Log-transformed variables created: log_stars, log_forks, log_issues")
```

## Section 3: Bivariate Analysis

Now examine **relationships between variables** — which pairs are correlated? Do group means differ?

### 3.1 Correlation Matrix

**Correlation** measures the strength and direction of the **linear relationship** between two variables, ranging from −1 to +1. A value near +1 indicates a strong positive relationship, −1 a strong negative relationship, and 0 indicates no linear relationship. Correlation does not imply causation.

```python
# Select the variables we plan to use in regression
analysis_vars = ["log_stars", "log_forks", "log_issues",
                 "desc_word_count", "topic_count", "repo_age_days",
                 "is_org", "is_python", "is_permissive", "mentions_ai"]

corr = df[analysis_vars].corr().round(2)

# Heatmap
fig, ax = plt.subplots(figsize=(10, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))  # Show lower triangle only
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap="RdBu_r",
            center=0, vmin=-1, vmax=1, square=True,
            linewidths=0.5, ax=ax)
ax.set_title("Correlation Matrix", fontsize=14, fontweight="bold", pad=15)
plt.tight_layout()
plt.show()
```

```python
# Flag high correlations (potential multicollinearity)
# Get pairs with |r| > 0.7 (excluding self-correlations)
high_corr = []
for i in range(len(corr.columns)):
    for j in range(i + 1, len(corr.columns)):
        r = corr.iloc[i, j]
        if abs(r) > 0.7:
            high_corr.append((corr.columns[i], corr.columns[j], r))

if high_corr:
    print("⚠️  High correlations (|r| > 0.7) — potential multicollinearity:")
    for v1, v2, r in high_corr:
        print(f"   {v1} × {v2}: r = {r}")
    print("\n   Consider: don't include both in the same regression model.")
else:
    print("✅ No variable pairs with |r| > 0.7. Multicollinearity is not a concern.")
```

## Section 4: Outlier Detection & Decisions

Outliers are observations that are far from the rest of the data. They can distort statistical results — but **not all outliers should be removed**.

<img src="outlier.png" width="80%">

```python
# Box plots to visualize outliers
fig, axes = plt.subplots(1, 3, figsize=(14, 5))

for ax, var in zip(axes, ["stargazers_count", "forks_count", "open_issues_count"]):
    ax.boxplot(df[var].dropna(), vert=True, patch_artist=True,
               boxprops=dict(facecolor="#DBEAFE"),
               medianprops=dict(color="#1B3A5C", linewidth=2))
    ax.set_title(var, fontweight="bold")
    ax.set_ylabel("Value")

plt.suptitle("Box Plots — Identifying Outliers", fontsize=13, fontweight="bold")
plt.tight_layout()
plt.show()
```

**IQR Method for Outlier Detection**

- The Interquartile Range (IQR) measures the spread of the middle 50% of the data: IQR = Q3 − Q1. Values falling below Q1 − 1.5 × IQR or above Q3 + 1.5 × IQR are flagged as outliers. 
- This method is robust to extreme values and does not assume a normal distribution.

```python
# IQR Method for outlier detection
def detect_outliers_iqr(series, multiplier=1.5):
    """Detect outliers using the IQR method."""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - multiplier * IQR
    upper = Q3 + multiplier * IQR
    outliers = series[(series < lower) | (series > upper)]
    return outliers, lower, upper

print("Outlier Detection (IQR method, 1.5× IQR):")
print("=" * 60)
for var in ["stargazers_count", "forks_count", "open_issues_count"]:
    outliers, lo, hi = detect_outliers_iqr(df[var].dropna())
    print(f"\n{var}:")
    print(f"  Bounds: [{lo:.0f}, {hi:.0f}]")
    print(f"  Outliers: {len(outliers)} rows")
    if len(outliers) > 0:
        print(f"  Examples: {outliers.head(5).tolist()}")
```

**Z-Score Method for Outlier Detection**

- The Z-score measures how many standard deviations a value is from the mean: Z = (X − μ) / σ. Values with |Z| > 3 (or 2) are typically flagged as outliers. 
- This method assumes an approximately normal distribution and is sensitive to extreme values, as both the mean and standard deviation can be distorted by outliers.

```python
# Z-score method
from scipy import stats

print("Outlier Detection (Z-score method, |z| > 3):")
print("=" * 60)
for var in ["stargazers_count", "forks_count", "open_issues_count"]:
    z = np.abs(stats.zscore(df[var].dropna()))
    n_outliers = (z > 3).sum()
    print(f"  {var}: {n_outliers} observations with |z| > 3")
```

### Deciding What to Do About Outliers

Not every outlier should be removed. The decision depends on **why** it's an outlier:

| Reason | Action | Example |
|---|---|---|
| Data entry error | Fix or remove | negative open_issues_count (already fixed in cleaning) |
| Genuine extreme value | Keep, but test robustness | DeepSeek-V3 with 685K stars — it's a real repo |
| Different population | Consider removing | Test repos with 0 stars (already removed in cleaning) |

**For this dataset:** We keep extreme values because they represent genuine variation in the open-source ecosystem. In the paper, we note: *"We retain all observations after data cleaning.
