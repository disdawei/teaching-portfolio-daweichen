```python
# Solution sketch for Exercise 3
try:
    user_input = input("Enter a number: ")
    x = int(user_input)
    result = 100 / x
except ValueError:
    print(f"Invalid input: '{user_input}' is not an integer.")
except ZeroDivisionError:
    print("Invalid input: division by zero is not allowed.")
else:
    print(f"Result is {result}")
```

```python
# Solution sketch for Exercise 4
def average_score(scores):
    print("scores:", scores)
    print("type:", type(scores))
    print("length:", len(scores))
    total = sum(scores)
    return total / len(scores)

print(average_score([80, 90, 100]))
```

```python
# Solution sketch for Exercise 5
def average_score(scores):
    assert isinstance(scores, list), "scores must be a list"
    assert len(scores) > 0, "scores cannot be empty"
    total = sum(scores)
    return total / len(scores)

print(average_score([70, 80, 90]))
```
